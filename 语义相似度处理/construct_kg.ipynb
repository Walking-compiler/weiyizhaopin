```python
import pandas as pd
import numpy as np
import re
from cal_similarity import compute_similarity
from write_file import write_1dlist_txt, write_2dlist_txt, write_entity2id_csv, write_entities2id_csv
import random
```

```python
kg_path = './new_kg.txt'
# 读取除LINK和MANAGER外的三元组
with open(kg_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()
    data = []
    for line in lines:
        row = line.strip()
        # 由于读入原因，去除除字母、数字和汉字外的所有字符
        new_row = re.sub(r'[^\u4e00-\u9fa5a-zA-Z0-9-—_\s]', '', row)
        new_row = new_row.split()
        if len(new_row) != 3:
            continue
        # 在构建适合输入KGCN的知识图谱时，不需要HR和LINK
        rel = new_row[1]
        if rel not in ['MANAGER', 'LINK']:
            data.append(new_row)
```

```python
print(len(data))
```

```python
new_data = []
for triplet in data:
    rel = triplet[1]
    if rel == 'LOCATE':
        triplet[2] = triplet[2][:2]
        new_data.append(triplet)
    elif rel == 'SALARY':
        numbers = re.findall(r'\d+', triplet[2])
        if len(numbers) not in [2, 3]:
            continue
        triplet_1 = [triplet[0], 'MIN_SALARY', numbers[0]]
        triplet_2 = [triplet[0], 'MAX_SALARY', numbers[1]]
        new_data.append(triplet_1)
        new_data.append(triplet_2)
        if len(numbers) == 3:
            triplet_3 = [triplet[0], 'is_13', 'True']
        else:
            triplet_3 = [triplet[0], 'is_13', 'False']
            new_data.append(triplet_3)
    else:
        new_data.append(triplet)
```

```python
for triplets in excerpt_data:
    if triplets[1] == 'REQUIRE':
        # 替换company
        for group in idx2company:
            if triplets[0] in group[0]:
                triplets[0] = f'c{group[1]}'

        # 替换job
        for group in idx2groupby_job:
            if triplets[2] in group[0]:
                triplets[2] = f'j{group[1]}'

    elif triplets[1] == 'LOCATE':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'

        # 替换location
        for group in idx2location:
            if triplets[2] in group[0]:
                triplets[2] = f'l{group[1]}'

    elif triplets[1] == 'PS':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'

        # 替换PS
        for group in idx2groupby_ps:
            if triplets[2] in group[0]:
                triplets[2] = f'ps{group[1]}'

    elif triplets[1] == 'JS':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'

        # 替换JS
        for group in idx2groupby_js:
            if triplets[2] in group[0]:
                triplets[2] = f'js{group[1]}'

    elif triplets[1] == 'JR':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'

        # 替换JR
        for group in idx2groupby_jr:
            if triplets[2] in group[0]:
                triplets[2] = f'jr{group[1]}'

    elif triplets[1] == 'JN':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'

        # 替换JN
        for group in idx2groupby_jn:
            if triplets[2] in group[0]:
                triplets[2] = f'jn{group[1]}'
    
    elif triplets[1] == 'EDUCATION':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'
            
        # 替换education
        for group in idx2education:
                if triplets[2] in group[0]:
                    triplets[2] = f'ed{group[1]}'
    
    elif triplets[1] == 'MIN_SALARY':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'
                
        # 替换min_salary
        for group in idx2min_salary:
            if triplets[2] in group[0]:
                triplets[2] = f'j{group[1]}'
    
    elif triplets[1] == 'MAX_SALARY':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'
                
        # 替换max_salary
        for group in idx2max_salary:
            if triplets[2] in group[0]:
                triplets[2] = f'j{group[1]}'
    
    elif triplets[1] == 'is_13':
        # 替换job
        for group in idx2groupby_job:
            if triplets[0] in group[0]:
                triplets[0] = f'j{group[1]}'
                
        # 替换is_13
        if triplets[2] == 'True':
            triplets[2] = f'1'
        else:
            triplets[2] = f'0'
    
```

```python
path = './to_change.csv'

df = pd.read_csv(path)

jobs_id = list(range(0, 48))

items_id = df['item_id'].tolist()

items_id = list(set(items_id))

items2jobs = {}

for index, _ in enumerate(items_id):
    replacement_index = random.randint(0, len(jobs_id) - 1)
    items2jobs[items_id[index]] = jobs_id[replacement_index]

df.replace(items2jobs, inplace=True)
df = df.drop_duplicates(subset=['item_id', 'user_id'], keep='first')
df = df.drop(columns='vtime')

action = {'click': 'clicked', 'collect': 'collected', 'cart': 'delivered', 'alipay': 'satisfy'}
df.replace(action, inplace=True)

df.to_csv('ori_u_i_action.csv', index=False, sep='\t')

# Convert four behaviors into explicit ratings
df.action = df.action.replace(
    ['clicked', 'collected', 'delivered', 'satisfy'],
    [1.25, 2.5, 3.75, 5])
df = df[['user_id', 'item_id', 'action']].rename({'action': 'rating'}, axis=1)

# Save to ratings.csv
df.to_csv('ratings.csv', index=False, sep='\t')
```
